```python
#Importing the datasets as csv files.
import pandas as pd

class DataReader:
  def read(self, file_name):
    raise NotImplementedError("Subclass must implement this method")

class CSVReader(DataReader):
  def read(self, file_name):
    return pd.read_csv(file_name)

def load_data(reader, file_name):
  return reader.read(file_name)

data_reader = CSVReader()
mat_file = load_data(data_reader, 'student-mat.csv')
por_file = load_data(data_reader, 'student-por.csv')
```

```python
#Create a Student class that encapsulates attributes and methods related to individual student records
class Student:
    def __init__(self, student_dict):
        self.data = student_dict

    def get_grade(self, subject):
        return self.data[f'G3_{subject}']

    def get_studytime(self, subject):
        return self.data[f'studytime_{subject}']
```

```python
#Design a DatasetManager class to handle loading, preprocessing, and transforming the dataset.
class DatasetManager:
    def __init__(self, math_file, por_file):
        self.math_file = math_file
        self.por_file = por_file
        self.students = []

    def load_data(self):
        df_math = pd.read_csv(self.math_file, sep=';')
        df_por = pd.read_csv(self.por_file, sep=';')
        first_columns = df_math.columns[:13]
        second_columns = df_math.columns[14:-1]
        new_list = first_columns.join(second_columns)
        merged_df = pd.merge(df_math, df_por, on=list(new_list), suffixes=('_math', '_por'), how='outer')    # Merge the DataFrames on all common attributes with an outer join
        merged_df.fillna({'G1_math': 0, 'G2_math': 0, 'G3_math': 0, 'G1_por': 0, 'G2_por': 0, 'G3_por': 0, 'studytime_math': 0,'studytime_por': 0}, inplace=True)   # Fill NaN values with 0 for grades
        student_profiles = merged_df.to_dict('records')   # Convert the merged DataFrame into a list of dictionaries
        for profile in student_profiles:
          self.students.append(Student(profile))

    def get_students(self):
        return self.students
```

```python
#Create a base class
class AnalysisBase:
    def analyze(self):
        raise NotImplementedError("Subclass must implement this method")
```

```python
#Develop subclasses that inherit from a base class, implementing specialized methods for different types of analysis (e.g., GradeAnalysis subclass for analyzing grades).
import numpy as np
import matplotlib.pyplot as plt

class GradeAnalysis_mean(AnalysisBase):   #Analyze the mean final grades
    def __init__(self, students):
        self.students = students

    def analyze(self, subject):
        grades = [student.get_grade(subject) for student in self.students]
        grades_filtered = [grade for grade in grades if grade != 0]   #filtered out values of 0 before calculating the mean
        mean_grade = np.mean(grades_filtered)
        return round(mean_grade, 2)

class GradeAnalysis_median(AnalysisBase):   #Analyze the median final grades
    def __init__(self, students):
        self.students = students

    def analyze(self, subject):
        grades = [student.get_grade(subject) for student in self.students]
        grades_filtered = [grade for grade in grades if grade != 0]   #filtered out values of 0 before calculating the median
        median_grade = np.median(grades_filtered)
        return round(median_grade,1)

class GradeAnalysis_StudyTime(AnalysisBase):   #Visualize the correlation between study time and final grades
    def __init__(self, students):
        self.students = students

    def analyze(self, subject):
        grades = [student.get_grade(subject) for student in self.students]
        study_time = [student.get_studytime(subject) for student in self.students]
        tuple_list = list(zip(study_time, grades))    #Create a list of tuples where each grade corresponds to a studytime value

grade_total_1 = 0
        count_1 = 0
        grade_total_2 = 0
        count_2 = 0
        grade_total_3 = 0
        count_3 = 0
        grade_total_4 = 0
        count_4 = 0
        for studytime_and_grade in tuple_list:
          studytime = studytime_and_grade[0]
          grade = studytime_and_grade[1]
          if studytime == 1.0:    #Calculate the average grade when studytime is a value of 1
            grade_total_1 += grade
            count_1 += 1
          elif studytime == 2.0:    #Calculate the average grade when studytime is a value of 2
            grade_total_2 += grade
            count_2 += 1
          elif studytime == 3.0:    #Calculate the average grade when studytime is a value of 3
            grade_total_3 += grade
            count_3 += 1
          elif studytime == 4.0:    #Calculate the average grade when studytime is a value of 4
            grade_total_4 += grade
            count_4 += 1

        avg_1 = grade_total_1 / count_1
        avg_2 = grade_total_2 / count_2
        avg_3 = grade_total_3 / count_3
        avg_4 = grade_total_4 / count_4

        avg_grades = [avg_1, avg_2, avg_3, avg_4]   #Combine all the averages into a new list to be graphed
        study_times = [1,2,3,4]   #Use the 4 possible values of studytime for the x-axis so that the number of x and y values are equal

        studytime_grade_plot = plt.bar(study_times, avg_grades)
        plt.xlabel("Study Time")
        plt.ylabel("Final Grade")
        plt.title("Study Time vs Final Grade")
        return studytime_grade_plot
```

```python
dataset_manager = DatasetManager('student-mat.csv', 'student-por.csv')
dataset_manager.load_data()
students = dataset_manager.get_students()

grade_analysis_mean = GradeAnalysis_mean(students)
grade_analysis_median = GradeAnalysis_median(students)
studytime_grade = GradeAnalysis_StudyTime(students)

#mean and median grades are analyzed by subject
mean_math_grade = grade_analysis_mean.analyze('math')
mean_por_grade = grade_analysis_mean.analyze('por')
median_math_grade = grade_analysis_median.analyze('math')
median_por_grade = grade_analysis_median.analyze('por')

print("Mean Math Grade:", mean_math_grade)
print("Median Math Grade:", median_math_grade)
print()
print("Mean Portuguese Grade:", mean_por_grade)
print("Median Portuguese Grade:", median_por_grade)
```
